## Importing my libaries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## loading my dataset from seaborn for visualization

```python
tips= sns.load_dataset('tips')

tips
```

## visualizing the Size of our Dataset

```python
sns.kdeplot(data=tips,shade=True)
plt.title('Size Feature of Tips Dataset')
plt.show()
```

Here we can see the size of the dataset and the amount of tips collected and the total_bills generated from the dataset

## Generating Total bill by gender for our tips dataset

```python
sns.boxplot(data=tips, x='sex', y='total_bill')
plt.title('Distribution of Total_bill by Gender')
plt.show()
```

 Here we can see that the male gender paid more bills than the female, judging by the maximumvalue which is at 40 for the male and at 30 for the female.

## Ploting for the distribution of tips by days

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.title('Distribution of Tips by Day')
plt.show()
```

Here we can see that sunday has the maximum tip givers on that day.

## Generating the total bill by day considering smoker status

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Distribution of Total_Bill by Day based on Smokers')
plt.show()
```

Here we can see that on other days we have almost the same values for smokers and non smokers bills,
for saturday we have more smokers than non smokers and it is the same for sunday.

## Generating a violin plot to show distribution of total bill by time based on gender




```python
sns.violinplot(data=tips, y='time', x='total_bill', hue='sex', orient='horizontal')
plt.title('Distribution of Total_bill by Time based on Gender')
plt.show()
```

Here we can see the distribution of the total_bill amount by time. Additionally, we split by gender. Immediately we see that the largest difference in the shape of the distribution between genders happens during dinner.

## Generating a violin plot to show distribution of tips by day based on time

```python
sns.violinplot(data=tips, y='day', x='tip', hue='time', orient='horizontal')
plt.title('Distribution of Tips by Day based on Time')
plt.show()
```

Here we can see the distribution of tips amount by day. Additionally, we split by time. Immediately we see that the largest difference in the shape of the distribution between time happens during dinner on saturday and sunday.

## Plotting and customizing a heatmap for the Tips Dataset

```python
ax = sns.heatmap(tips.corr(), annot=True, fmt='.1f', cmap='Greens', linewidths=8, cbar_kws={'shrink':.99})
plt.title('Correlations Heatmap of Tips Dataset \n'.title())
plt.show()
```
