### Importing my libaries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Inspecting and Loading my dataset into a dataframe(df)

```python
df= pd.read_csv('WorldCupMatches[1].csv')
df.head(10)
```

# Inspecting my Dataframe dtypes

```python
df.dtypes
```

## checking for missing values in my dataframe

```python
df.isnull().sum()
```

## Droping missing values and empty rows to prepare my dataframe for visualization 

```python
df.dropna( inplace = True)
df
```

## Creating a new column in df named Total Goals

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head(10)
```

## changing the datatype of my year column

```python
df['Year']= df['Year'].astype(int)
```

## Visualize the total number of goals scored in each match.

```python
sns.displot(df['Total Goals'] , bins = 20, kde=False)
plt.title('Total Number of Goals in each Match')
plt.show()
```

##  Goals scored between 1930 - 2014

```python
sns.set_style('whitegrid')
sns.set_context(rc={'font.size': 10, 'xtick.labelsize':8})
sns.barplot(data=df, x='Year', y='Total Goals', ci=None, estimator=sum)
plt.xticks(rotation='vertical')
plt.title('Total Number of Goals Scored Yearly')
plt.show()
```

## creating a subplot with 12inches wide and 7 inches tall for my previous plot

```python
fig= plt.subplots(figsize=(12,7))
ax = sns.barplot(data=df, x="Year", y="Total Goals", ci=None, estimator=sum)
ax.set_title("Total Goals Yearly")  
plt.show()
```

## visualizing columns of year and total goals

```python
ax = sns.barplot(data=df,x='Year', y='Total Goals', ci=None )
ax.set_title('World Cup Goals')
plt.xticks(rotation='vertical')
plt.show()
```

## creating a box plot to show the average goals

```python
fig= plt.figure(figsize=(12,7))
sns.boxplot(data=df,x='Year', y='Total Goals')
plt.title('Total Number of Goals on a Average');
```
